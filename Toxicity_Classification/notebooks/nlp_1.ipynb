```python
import os 
import numpy as np
trainX = np.array([])
labels = []
path = raw_data_path
authors = os.listdir(path)
authors = [i for i in authors if '.DS_Store' not in i]
```

```python
for auth in authors:
    if authors != '.DS_Store':
        files = os.listdir(path+'\\'+auth+'\\')
        tmpX, tmpY = np.array([]),[]
        for file in files:
            f = open(path +'\\'+ auth+'\\'+file,'r')
            data = f.read().replace('\n',' ')
            tmpX = np.append(tmpX, data)
            tmpY = tmpY + [auth]
            f.close()
    trainX = np.append(trainX, tmpX)
    labels = labels+ tmpY
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(trainX, labels, train_size=0.8)
X_train[1]
```

**Exercise 2:**

1. Use word_tokenize to split flles into tokens
2. Find the number of tokens & unique tokens for each file
3. Find the average tokens & unique tokens per author
4. Return top and bottom 2 files based on number of tokens

```python
def read_file_data(filePath):
    f = open(filePath,'r')        
    data = f.read().replace('\n',' ')
    return data
```

```python
def tokenize_data(data):
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(data)
    return tokens
    
```

```python
def token_lens(tokens):
    token_len = len(tokens)
    unique_token_len = len(set(tokens))
    return token_len, unique_token_len
```

```python
auth_dict={'authors':[],'avg_tokens':[],'avg_unique_tokens':[]}
file_dict={'authors':[],'files':[],"total_counts":[],"unique_counts":[]}
for auth in authors:
    files = os.listdir(path+'\\'+auth+'\\')
    auth_dict['authors'].append(auth)
    token_count=0
    unique_token_count=0
    file_count=0
    for file in files:
        data = read_file_data(path+'\\'+auth+'\\'+file)
        tokens = tokenize_data(data)
        token_len,unique_len = token_lens(tokens)
        file_dict['authors'].append(auth)
        file_dict['files'].append(file)
        file_dict['total_counts'].append(token_len)
        file_dict['unique_counts'].append(unique_len)
        token_count+=token_len
        unique_token_count+=unique_len
        file_count+=1
    auth_dict['avg_tokens'].append(token_count/file_count)
    auth_dict['avg_unique_tokens'].append(unique_token_count/file_count)
```

```python
auth_df = pd.DataFrame(auth_dict)
file_df = pd.DataFrame(file_dict)
```
